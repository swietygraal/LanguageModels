```python
from transformers import AutoTokenizer, AutoModel
import torch
import random
import numpy as np

# ===== 1) BERT jako ekstraktor cech (jak w starterze) =====
model_name = "allegro/herbert-base-cased"
device = "cuda"  # albo 'cpu'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name).to(device)
model.eval()
```

```python
def representation(words):
    txt = " ".join(words)
    inp = tokenizer(txt, return_tensors="pt", truncation=True, max_length=128).to(device)
    with torch.no_grad():
        out = model(**inp)
    # [CLS]
    return out.last_hidden_state[0, 0].detach().cpu().numpy()


# ===== 2) Augmentacja mechaniczna (niedeterministyczna) =====
PL2LAT = str.maketrans({
    "ą":"a","ć":"c","ę":"e","ł":"l","ń":"n","ó":"o","ś":"s","ż":"z","ź":"z",
    "Ą":"A","Ć":"C","Ę":"E","Ł":"L","Ń":"N","Ó":"O","Ś":"S","Ż":"Z","Ź":"Z"
})

# bardzo uproszczone „sąsiedztwo” klawiatury PL (QWERTY) – wystarcza do literówek
NEIGH = {
    "a":"sqwz", "s":"awedxz", "d":"serfcx", "f":"drtgvc", "g":"ftyhbv", "h":"gyujnb", "j":"huikmn",
    "k":"jiolm", "l":"kop",
    "q":"wa", "w":"qase", "e":"wsdr", "r":"edft", "t":"rfgy", "y":"tghu", "u":"yhji", "i":"ujko", "o":"iklp", "p":"ol",
    "z":"asx", "x":"zsdc", "c":"xdfv", "v":"cfgb", "b":"vghn", "n":"bhjm", "m":"njk",
}

def random_case(s, p=0.08):
    # losowo podbije literę do upper / lower
    chars = list(s)
    for i, ch in enumerate(chars):
        if ch.isalpha() and random.random() < p:
            chars[i] = ch.upper() if random.random() < 0.5 else ch.lower()
    return "".join(chars)

def remove_diacritics(s, p=0.30):
    # z pewnym prawdopodobieństwem usuń polskie znaki w całym tekście
    return s.translate(PL2LAT) if random.random() < p else s

def typo_char_substitution(s, p=0.05):
    chars = list(s)
    for i, ch in enumerate(chars):
        low = ch.lower()
        if low in NEIGH and random.random() < p:
            repl = random.choice(NEIGH[low])
            chars[i] = repl.upper() if ch.isupper() else repl
    return "".join(chars)

def delete_random_char(s, p=0.03):
    if len(s) < 4 or random.random() > p:
        return s
    i = random.randrange(len(s))
    return s[:i] + s[i+1:]

def duplicate_random_char(s, p=0.03):
    if len(s) < 4 or random.random() > p:
        return s
    i = random.randrange(len(s))
    return s[:i] + s[i] + s[i:]

def swap_adjacent_chars(s, p=0.03):
    if len(s) < 4 or random.random() > p:
        return s
    i = random.randrange(len(s)-1)
    if s[i].isspace() or s[i+1].isspace():
        return s
    return s[:i] + s[i+1] + s[i] + s[i+2:]

def mess_punctuation(s, p=0.10):
    if random.random() > p:
        return s
    # usuń losowo część kropek/przecinków albo dodaj wykrzyknik
    if random.random() < 0.5:
        return s.replace(",", "", 1) if "," in s else s.replace(".", "", 1)
    else:
        return s + random.choice(["!", "!!", "!?"])

def augment_text_mechanical(text):
    # pipeline niedeterministycznych zakłóceń
    t = text
    #t = remove_diacritics(t, p=0.35)
    t = random_case(t, p=0.10)
    #t = typo_char_substitution(t, p=0.06)
    t = delete_random_char(t, p=0.05)
    t = duplicate_random_char(t, p=0.04)
    #t = swap_adjacent_chars(t, p=0.04)
    #t = mess_punctuation(t, p=0.12)
    return t

def augment_words(words):
    # words -> text -> augment -> split
    txt = " ".join(words)
    aug = augment_text_mechanical(txt)
    return aug.split()
```

```python
# ===== 3) Wczytanie danych i podział =====
lines = open("reviews_for_task3.txt", encoding="utf-8").readlines()
random.shuffle(lines)

N = len(lines)
test_size = N // 4
train_size = N - test_size

train_lines = lines[:train_size]
test_lines  = lines[train_size:]
```

```python
# ===== 4) Budowa datasetu z augmentacją K =====
def build_dataset(train_lines, test_lines, K=3):
    X_train, y_train = [], []
    X_test, y_test = [], []

    # train
    for line in train_lines:
        L = line.strip().split()
        y = 0 if L[0] == "BAD" else 1
        base_words = L[1:]

        # oryginał
        X_train.append(representation(base_words))
        y_train.append(y)

        # K augmentacji (niedeterministycznie)
        for _ in range(K):
            aug_words = augment_words(base_words)
            X_train.append(representation(aug_words))
            y_train.append(y)

        if len(X_train) % 200 == 0:
            print("train feats:", len(X_train))

    # test bez augmentacji
    for line in test_lines:
        L = line.strip().split()
        y = 0 if L[0] == "BAD" else 1
        base_words = L[1:]
        X_test.append(representation(base_words))
        y_test.append(y)

        if len(X_test) % 200 == 0:
            print("test feats:", len(X_test))

    return np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test)
```

```python
# ===== 5) Uczenie: regresja logistyczna =====
from sklearn.linear_model import LogisticRegression

def train_and_eval(K):
    X_train, y_train, X_test, y_test = build_dataset(train_lines, test_lines, K=K)

    clf = LogisticRegression(max_iter=2000, random_state=0)

    clf.fit(X_train, y_train)
    tr = clf.score(X_train, y_train)
    te = clf.score(X_test, y_test)
    return tr, te
```

```python
import random
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel, AutoModelForCausalLM

# ===== BERT jako ekstraktor cech =====
bert_name = "allegro/herbert-base-cased"
device = "cuda"  # albo "cpu"

bert_tokenizer = AutoTokenizer.from_pretrained(bert_name)
bert_model = AutoModel.from_pretrained(bert_name).to(device)
bert_model.eval()

def representation_from_text(txt: str):
    inp = bert_tokenizer(txt, return_tensors="pt").to(device)
    with torch.no_grad():
        out = bert_model(**inp)
    return out.last_hidden_state[0, 0].detach().cpu().numpy()

# ===== Model generatywny (Polka/Papuga/Polish GPT2) =====
# Podmień na właściwy model Polka/Papuga, albo zostaw lokalną ścieżkę jak wcześniej
gen_name = "eryk-mazus/polka-1.1b"  # np. Polka/Papuga
gen_tokenizer = AutoTokenizer.from_pretrained(gen_name)
gen_model = AutoModelForCausalLM.from_pretrained(gen_name).to(device)
gen_model.eval()

# GPT-2 czasem nie ma pad_token -> ustawiamy na eos_token
if gen_tokenizer.pad_token_id is None:
    gen_tokenizer.pad_token = gen_tokenizer.eos_token
```

```python
def build_prompt(review: str, label: str) -> str:
    sent = "pozytywna" if label == "GOOD" else "negatywna"
    # Bardzo prosty prompt; w razie czego możesz go stroić
    return (
        f"Zadanie: Napisz krótką recenzję w języku polskim, zachowując ten sam sentyment: {sent}.\n"
        f"Możesz zmienić szczegóły i wymyślić nowe, ale ton ma pozostać {sent}.\n"
        f"Oryginał: {review}\n"
        f"Nowa recenzja ({sent}):"
    )
```

```python
def generate_one(review: str, label: str,
                 max_new_tokens=60, top_p=0.92, temperature=0.95, repetition_penalty=1.1):
    prompt = build_prompt(review, label)

    enc = gen_tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        out = gen_model.generate(
            **enc,
            do_sample=True,
            top_p=top_p,
            temperature=temperature,
            repetition_penalty=repetition_penalty,
            max_new_tokens=max_new_tokens,
            pad_token_id=gen_tokenizer.pad_token_id,
            eos_token_id=gen_tokenizer.eos_token_id,
        )

    decoded = gen_tokenizer.decode(out[0], skip_special_tokens=True)

    # wytnij prompt, zostaw tylko wygenerowaną część
    gen_part = decoded[len(prompt):].strip()
    # czasem model „dopisuje instrukcje” — obcinamy na pierwszym pustym wierszu itp.
    gen_part = gen_part.split("\n")[0].strip()

    # fallback gdyby wyszło pusto
    if len(gen_part) < 5:
        gen_part = review

    return gen_part


def generate_k(review: str, label: str, K: int):
    return [generate_one(review, label) for _ in range(K)]
```

```python
lines = open("reviews_for_task3.txt", encoding="utf-8").read().splitlines()
random.shuffle(lines)

N = len(lines)
test_size = N // 4
train_lines = lines[:-test_size]
test_lines  = lines[-test_size:]


def parse_line(line: str):
    first_space = line.find(" ")
    label = line[:first_space]
    review = line[first_space+1:].strip()
    return label, review


def build_dataset_generative(train_lines, test_lines, K=3):
    X_train, y_train = [], []
    X_test, y_test = [], []

    # TRAIN: oryginał + K generacji
    for idx, line in enumerate(train_lines, 1):
        label, review = parse_line(line)
        y = 1 if label == "GOOD" else 0

        # oryginał
        X_train.append(representation_from_text(review))
        y_train.append(y)

        # augmentacje generatywne
        gens = generate_k(review, label, K)
        for g in gens:
            X_train.append(representation_from_text(g))
            y_train.append(y)

        if idx % 50 == 0:
            print("train done:", idx)

    # TEST: bez augmentacji
    for idx, line in enumerate(test_lines, 1):
        label, review = parse_line(line)
        y = 1 if label == "GOOD" else 0
        X_test.append(representation_from_text(review))
        y_test.append(y)
        if idx % 100 == 0:
            print("test done:", idx)

    return np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test)
```

```python
from sklearn.linear_model import LogisticRegression

def train_and_eval_generative(K):
    X_train, y_train, X_test, y_test = build_dataset_generative(train_lines, test_lines, K=K)

    clf = LogisticRegression(max_iter=2000, random_state=0)

    clf.fit(X_train, y_train)
    return clf.score(X_train, y_train), clf.score(X_test, y_test)


for K in [0, 1, 2, 3, 4, 5]:
    tr, te = train_and_eval_generative(K)
    print(f"K={K:>2} | Train acc={tr:.4f} | Test acc={te:.4f}")
```

```python
import re
import random
from collections import defaultdict

def load_superbazy(path="superbazy.txt"):
    form2lemma = {}
    lemma2forms = defaultdict(list)
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            form, lemma = line.split()
            form2lemma[form] = lemma
            lemma2forms[lemma].append(form)
    return form2lemma, lemma2forms

form2lemma, lemma2forms = load_superbazy("superbazy.txt")
```

```python
from gensim.models import Word2Vec

def tokenize_text(text):
    # proste tokeny: słowa i znaki interpunkcyjne jako osobne tokeny
    return re.findall(r"\w+|[^\w\s]", text, flags=re.UNICODE)

def lemmatize_tokens(tokens):
    out = []
    for t in tokens:
        # zostaw interpunkcję
        if re.fullmatch(r"[^\w\s]", t):
            out.append(t)
            continue
        # lemat z superbazy lub lowercase jako fallback
        out.append(form2lemma.get(t, t.lower()))
    return out

# wczytaj recenzje
lines = open("reviews_for_task3.txt", encoding="utf-8").read().splitlines()
random.shuffle(lines)

N = len(lines)
test_size = N // 4
train_lines = lines[:-test_size]
test_lines  = lines[-test_size:]

texts = []
for line in train_lines:
    first_space = line.find(" ")
    review = line[first_space+1:].strip()
    toks = tokenize_text(review)
    lemmas = lemmatize_tokens(toks)
    texts.append(lemmas)

# trenuj Word2Vec
w2v = Word2Vec(
    sentences=texts,
    vector_size=200,
    window=5,
    min_count=2,
    workers=4,
    sg=1,           # skip-gram
    negative=10,
    epochs=10
)
wv = w2v.wv
```

```python
STOP_LIKE = {
    "i","oraz","a","ale","że","to","w","we","na","do","z","za","od","po","nie","tak","jak",
    "się","jest","było","była","byli","były","ma","mam","mamy","są"
}

def is_word(token):
    return re.fullmatch(r"\w+", token) is not None

def is_replaceable(token):
    if not is_word(token):
        return False
    if len(token) < 4:
        return False
    if token.lower() in STOP_LIKE:
        return False
    if re.search(r"\d", token):
        return False
    return True

def best_form_by_suffix(original_form, candidate_forms):
    # próbuje dopasować końcówki od najdłuższych
    orig = original_form.lower()
    cands = candidate_forms

    # długości końcówek do sprawdzenia
    for L in [5,4,3,2,1]:
        suf = orig[-L:] if len(orig) >= L else orig
        matches = [f for f in cands if f.lower().endswith(suf)]
        if matches:
            return random.choice(matches)

    # jeśli nic nie pasuje — wybierz dowolną formę
    return random.choice(cands) if cands else None

def keep_capitalization(original, new):
    if original.isupper():
        return new.upper()
    if original[0].isupper():
        return new[0].upper() + new[1:]
    return new
```

```python
def replace_token_w2v(token, topn=25):
    # jeśli token nie jest w superbazie -> trudniej zachować formę -> często lepiej pominąć
    if token not in form2lemma:
        return token

    lemma = form2lemma[token]

    if lemma not in wv:
        return token

    # kandydaci-lemmaty
    candidates = [w for (w, sim) in wv.most_similar(lemma, topn=topn)]
    random.shuffle(candidates)

    for cand_lemma in candidates:
        # musi mieć formy w superbazie
        forms = lemma2forms.get(cand_lemma, [])
        if not forms:
            continue

        new_form = best_form_by_suffix(token, forms)
        if not new_form:
            continue

        new_form = keep_capitalization(token, new_form)

        # nie zwracaj identycznego
        if new_form.lower() == token.lower():
            continue

        return new_form

    return token

def augment_w2v_tokens(tokens, replace_prob=0.20):
    out = []
    for t in tokens:
        if is_replaceable(t) and random.random() < replace_prob:
            out.append(replace_token_w2v(t))
        else:
            out.append(t)
    return out
```

```python
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel

# ===== BERT feature extractor (jak wcześniej) =====
bert_name = "allegro/herbert-base-cased"
device = "cuda"  # lub "cpu"

tokenizer = AutoTokenizer.from_pretrained(bert_name)
model = AutoModel.from_pretrained(bert_name).to(device)
model.eval()

def representation(words):
    txt = " ".join(words)
    inp = tokenizer(txt, return_tensors="pt").to(device)
    with torch.no_grad():
        out = model(**inp)
    return out.last_hidden_state[0, 0].detach().cpu().numpy()
```

```python
# ===== split train/test =====
import random
lines = open("reviews_for_task3.txt", encoding="utf-8").readlines()
random.shuffle(lines)

N = len(lines)
test_size = N // 4
train_lines = lines[:-test_size]
test_lines  = lines[-test_size:]

def build_dataset_word2vec(train_lines, test_lines, K=3, replace_prob=0.20):
    X_train, y_train = [], []
    X_test, y_test = [], []

    # TRAIN
    for i, line in enumerate(train_lines, 1):
        L = line.strip().split()
        y = 0 if L[0] == "BAD" else 1
        base = L[1:]

        # oryginał
        X_train.append(representation(base))
        y_train.append(y)

        # K augmentacji
        for _ in range(K):
            aug = augment_w2v_tokens(base, replace_prob=replace_prob)
            X_train.append(representation(aug))
            y_train.append(y)

        if i % 100 == 0:
            print("train lines:", i)

    # TEST bez augmentacji
    for i, line in enumerate(test_lines, 1):
        L = line.strip().split()
        y = 0 if L[0] == "BAD" else 1
        base = L[1:]
        X_test.append(representation(base))
        y_test.append(y)
        if i % 200 == 0:
            print("test lines:", i)

    return np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test)
```

```python
from sklearn.linear_model import LogisticRegression

def train_and_eval_word2vec(K, replace_prob=0.20):
    X_train, y_train, X_test, y_test = build_dataset_word2vec(train_lines, test_lines, K=K, replace_prob=replace_prob)

    clf = LogisticRegression(max_iter=2000, random_state=0)
    clf.fit(X_train, y_train)

    tr = clf.score(X_train, y_train)
    te = clf.score(X_test, y_test)
    return tr, te

for K in [0, 1, 2, 3, 4, 5]:
    tr, te = train_and_eval_word2vec(K=K, replace_prob=0.20)
    print(f"K={K:>2} | Train acc={tr:.4f} | Test acc={te:.4f}")
```
